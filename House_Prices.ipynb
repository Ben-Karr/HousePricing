## Load data

```python
path = Path('data')
path.mkdir(parents=True, exist_ok=True)
```

```python
train_df = pd.read_csv(path / 'train.csv')
test_df = pd.read_csv(path / 'test.csv')
```

## Combine Train & Test set and do preprocessing

```python
train_df['is_test'] = False

test_df['SalePrice'] = np.nan
test_df['is_test'] = True

df_all = pd.concat([train_df,test_df])
```

```python
df_all.tail()
```

```python
dep_var = 'SalePrice' # target

cat_names = ['MSSubClass','MSZoning','Street','Alley','LotShape','LandContour',
             'Utilities','LotConfig','LandSlope','Neighborhood','Condition1',
            'Condition2','BldgType','HouseStyle','OverallQual','OverallCond',
            'RoofStyle','RoofMatl','Exterior1st','Exterior2nd','MasVnrType',
            'ExterQual','ExterCond','Foundation','BsmtQual','BsmtCond',
            'BsmtExposure','BsmtFinType1','BsmtFinType2','Heating','HeatingQC',
            'CentralAir','Electrical','BsmtFullBath','BsmtHalfBath','FullBath',
            'HalfBath','KitchenQual','TotRmsAbvGrd',
            'Functional','Fireplaces','FireplaceQu','GarageType','GarageYrBlt',
            'GarageFinish','GarageCars','GarageQual','GarageCond','PavedDrive',
            'PoolQC','Fence','MiscFeature','MoSold','YrSold','SaleType',
             'SaleCondition','YearBuilt','YearRemodAdd','BedroomAbvGr', 'KitchenAbvGr']

cont_names = ['LotFrontage','LotArea','MasVnrArea',
              'BsmtFinSF1','BsmtFinSF2','BsmtUnfSF','TotalBsmtSF','1stFlrSF',
             '2ndFlrSF','LowQualFinSF','GrLivArea','GarageArea','WoodDeckSF',
             'OpenPorchSF','EnclosedPorch','3SsnPorch','ScreenPorch','PoolArea',
             'MiscVal']

procs = [FillMissing, Categorify, Normalize]
```

```python
categorify = Categorify(cat_names, cont_names)
categorify(df_all)

normalize = Normalize(cat_names, cont_names)
normalize(df_all)

fill_miss = FillMissing(cat_names, cont_names)
fill_miss(df_all)
```

```python
cat_na = [x for x in df_all.columns.tolist() if '_na' in x]
cat_names.extend(cat_na)

df_all[cat_na] = df_all[cat_na].astype('category')
```

```python
df_all.head()
```

## Split dataframes back to Train & Test

```python
train_df = df_all[df_all['is_test'] == False].drop(['is_test'],axis = 1)
test_df = df_all[df_all['is_test'] == True].drop(['is_test','SalePrice'],axis =1)
```

## Build DataBunch

```python
test = TabularList.from_df(test_df, path = path, cat_names = cat_names, cont_names = cont_names)
```

```python
data = (TabularList.from_df(train_df, path = path, cat_names = cat_names, cont_names = cont_names)
        .split_by_rand_pct(0.2, seed = 41)
        .label_from_df(cols = dep_var, label_cls=FloatList, log = True) # Predict the log of the target variable
        .add_test(test)
        .databunch())
```

```python
data.show_batch()
```

We are training to predict the log of the SalePrice.

```python
min_y = np.log(train_df['SalePrice']).min()
max_y = np.log(train_df['SalePrice']).max()
y_range = [min_y * 0.8, max_y * 1.2]
```

## Building the learner

```python
lyrs = [1000, 500]
ps = [0.001,0.01]
```

```python
learn = tabular_learner(data, 
                        layers = lyrs, 
                        metrics = rmse, 
                        ps = ps, 
                        y_range = y_range)
```

## Learn

```python
learn.lr_find()
learn.recorder.plot(suggestion = True)
```

```python
lr=1e-4
learn.fit_one_cycle(10,lr)
```

```python
lr = lr / 3
learn.fit_one_cycle(10,lr)
```

```python
learn.lr_find()
learn.recorder.plot(suggestion = True)
```

```python
lr = 1e-6
learn.fit_one_cycle(5, slice(lr))
```

## Predict the labels of the Test set

```python
predictions, _ = learn.get_preds(DatasetType.Test)
predictions = np.array(predictions[:,0])
```

## Save the submission
Since we predicted the log, we have to save $e^{prediction}$.

```python
submission = pd.DataFrame({'Id': test_df['Id'], 'SalePrice': np.exp(predictions)})
submission.to_csv('submission_pdPreproc.csv', index = False)
submission.head()
```
